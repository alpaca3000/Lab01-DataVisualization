```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('../Data/StudentPerformanceFactors.csv')
```

#### Câu 1: Liệu có mối tương quan nào giữa giới tính và thành tích học tập?

```python
plt.figure(figsize=(14, 10))

# 1. Biểu đồ mật độ: Giới tính và điểm số trước đó
plt.subplot(2, 1, 1)
sns.kdeplot(data=df[df['Gender'] == 'Male'], x='Previous_Scores', fill=True, common_norm=False,
             color='blue', alpha=0.3, label='Nam')
sns.kdeplot(data=df[df['Gender'] == 'Female'], x='Previous_Scores', fill=True, common_norm=False,
             color='yellow', alpha=0.4, label='Nữ')
plt.title('Mật độ điểm số trước đó theo giới tính', fontsize=16)
plt.xlabel('Điểm số trước đó', fontsize=14)
plt.ylabel('Mật độ', fontsize=14)
plt.legend()

# 2. Biểu đồ mật độ: Giới tính và điểm thi cuối kỳ
plt.subplot(2, 1, 2)
sns.kdeplot(data=df[df['Gender'] == 'Male'], x='Exam_Score', fill=True, common_norm=False,
             color='blue', alpha=0.3, label='Nam')
sns.kdeplot(data=df[df['Gender'] == 'Female'], x='Exam_Score', fill=True, common_norm=False,
             color='yellow', alpha=0.4, label='Nữ')
plt.title('Mật độ điểm thi cuối kỳ theo giới tính', fontsize=16)
plt.xlabel('Điểm thi cuối kỳ', fontsize=14)
plt.ylabel('Mật độ', fontsize=14)
plt.legend()

plt.tight_layout()
plt.show()
```

- Kết luận:
    - Biểu đồ mật độ cho thấy sự phân bố điểm số của cả nam và nữ khá tương đồng. Điều này cho thấy không có sự khác biệt rõ rệt trong thành tích học tập giữa hai giới tính.
    - Kết quả này cho thấy rằng giới tính không phải là yếu tố quyết định đáng kể đến thành tích học tập. Các yếu tố khác như chất lượng giáo viên, và điều kiện học tập có thể có ảnh hưởng lớn hơn đến điểm số của học sinh.

#### Câu 2: Giấc ngủ và các hoạt động thể chất ảnh hưởng như thế nào đến thành tích học tập?

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Sleep_Hours', y='Exam_Score', hue='Physical_Activity', palette='viridis', alpha=0.7)
plt.title('Điểm thi cuối kỳ theo số giờ ngủ và hoạt động thể chất', fontsize=16)
plt.xlabel('Số giờ ngủ')
plt.ylabel('Điểm thi cuối kỳ')
plt.legend(title='Mức độ hoạt động thể chất', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Physical_Activity', y='Exam_Score', color='lightblue')  # dùng color thay cho palette
plt.title('Điểm thi theo mức độ hoạt động thể chất', fontsize=16)
plt.xlabel('Mức độ hoạt động thể chất')
plt.ylabel('Điểm thi cuối kỳ')
plt.tight_layout()
plt.show()
```

```python
variables = ['Sleep_Hours', 'Exam_Score', 'Physical_Activity']

g = sns.PairGrid(df[variables])
g.map_diag(sns.histplot, kde=True, color='blue')
g.map_upper(sns.scatterplot, color='orange')
g.map_lower(sns.scatterplot, color='green')

plt.suptitle('Correlogram giữa Số giờ ngủ, Điểm thi và Mức độ hoạt động thể chất', y=1.02, fontsize=16)

plt.show()
```
